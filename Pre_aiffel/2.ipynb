```python
name = ['Hoka Clifton 8', 'Brooks Adrenaline GTS 22', 'Ryka Women\'s Devotion Plus 4 Walking Shoe',
        'Nike Women\'s Pegasus 39', 'Asics Gel-Excite 9', 'Hoka Women\'s Bondi 8',
        'Asics Women\'s GT-2000 10 Running Shoes']
material = ['Mesh', 'Recycled plastic bottle',
            'mesh', 'MESH', 'Mesh', 'Mesh', 'Knit']
weight = ['7.2 ounces', '10.2 ounces', '7.4 ounces',
          '9.6 ounces', '7.7 ounces', '8.9 ounces', '8.4 ounces']
price = ['$140', '$132', '$92', '$130', '$75', '$165', '$130']
```

1. `material` 리스트의 원소들을 소문자로 변경하기
    - 소재는 크게 3종류의 데이터로 이루어져 있습니다.(범주형 변수) 하지만 메쉬 소재의 경우 `mesh` , `MESH` , `Mesh` 로 다양하게 표현되고 있습니다. 3가지의 표기법을 통일시켜 주어야 하겠네요!

```python
import pandas as pd

dic = {}
dic['name'] = name
dic['material'] = material
dic['weight'] = weight
dic['price'] = price

df = pd.DataFrame.from_dict(dic)
df.info()
```

```python
df.head(10)
```

```python
df.material = df.material.apply(lambda x: x.lower())
df.material = df.material.astype('category')
df
```

2. `weight` 리스트의 ‘ounces’ 제거하고 숫자만 남기기
    - 무게 정보는 숫자로 이루어져 있습니다. (연속형 변수) 하지만 수집한 데이터에서는 ‘ounces’라는 단위가 붙어 숫자 자료형으로 취급할 수 없는 상태입니다. ‘ounces’를 제거하고 숫자만 남겨서 다루기 쉽게 변형시켜 봅시다.

```python
df.weight = df.weight.apply(lambda x: float(x.split()[0]))
df
```

3.  `price` 리스트의 ‘$’ 제거하고 숫자만 남기기
    - 가격 정보도 마찬가지로 연속형 변수입니다. ‘$’를 제거하고 숫자만 남겨서 다루기 쉽게 변형시켜 봅시다.

```python
df.price = df.price.apply(lambda x: int(x[1:]))
df
```

4. `brand` 리스트 만들기
    - 수집한 데이터로부터 다양한 데이터의 표현을 이끌어내는 것도 때로는 중요합니다. 현재 데이터에서는 이름 정보로부터 신발 브랜드를 추측할 수 있을 것 같습니다. 이름의 첫 번째 단어만 가지고 `brand` 라는 새로운 정보를 만들어 봅시다.

```python
df['brand'] = df.name.apply(lambda x: x.split()[0])
df
```
